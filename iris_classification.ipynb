```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris= pd.read_csv("iris.csv") #Load Iris.csv into a pandas dataframe
print(iris.shape) #Data-point and features
print(iris.columns)
```

```python
iris["variety"].value_counts()
```

```python
iris.plot(kind='scatter',x='sepal.length',y='sepal.width');
plt.show()
```

```python
sns.set_style('whitegrid')
sns.FacetGrid(iris, hue='variety',height=5) \
   .map(plt.scatter,'sepal.length','sepal.width')\
   .add_legend()
plt.show()
```

```python
#From above plot, Versicolor and Virginica cannot be distinguish
```

```python
sns.set_style("whitegrid")
sns.pairplot(iris, hue="variety", height=3);
plt.show()
```

```python
sns.FacetGrid(iris, hue="variety", size=5) \
.map(sns.distplot, "petal.length") \
.add_legend();
plt.show();
```

```python
sns.FacetGrid(iris,hue="variety",size=5)\
.map(sns.distplot,"petal.width")\
.add_legend();
plt.show()
print("Hello")
```
